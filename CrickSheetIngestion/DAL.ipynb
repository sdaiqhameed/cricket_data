```python
class DAL:
    __instance = None
    
    @staticmethod 
    def getInstance():
        """ Static access method. """
        if DAL.__instance == None:
            DAL()
        return DAL.__instance
    
    def __init__(self):
        if DAL.__instance != None:
            raise Exception("This class is a singleton!")
        else:
            DAL.__instance = self

    def get_connection_string(self):
        try:
            # DRIVER_NAME = "SQL SERVER"
            # SERVER_NAME = "acadia2.c2mek2byh6su.us-east-2.rds.amazonaws.com"
            # DATABASE_NAME = "Takhti20"
            # UID = "takhtiadmin"
            # PWD = "wQk8#RExL"

            DRIVER_NAME = "SQL SERVER"
            SERVER_NAME = "your server name" #"BLACK\\SQLEXPRESS"
            DATABASE_NAME = "Cricket"
            UID = "your user name"
            PWD = "your password"
            CONNECTION_STRING = f"DRIVER={DRIVER_NAME}; SERVER={SERVER_NAME}; DATABASE={DATABASE_NAME}; uid={UID}; pwd={PWD};"
            return CONNECTION_STRING
        except Exception as ex:
            return None
        
    def connect(self):
        try:
            CONNECTION_STRING = self.get_connection_string()
            if CONNECTION_STRING is not None:
                conn = odbc.connect(CONNECTION_STRING)
                return conn
            return None
        except Exception as ex:
            print(ex)
            return None
    
    def select_all(self, query):
        conn = self.connect()
        try:
            crsr = conn.cursor()
            crsr.execute(query)
            rows = crsr.fetchall()
            crsr.close()
            conn.close()
            return rows
        except Exception as ex:
            conn.close()
            return None
        
    def select_one_record(self, query):
        conn = self.connect()
        try:
            crsr = conn.cursor()
            crsr.execute(query)
            rows = crsr.fetchone()
            crsr.close()
            conn.close()
            return rows
        except Exception as ex:
            conn.close()
            return None
    
    def select_one_record_with_params(self, query, params):
        conn = self.connect()
        try:
            crsr = conn.cursor()
            crsr.execute(query, params)
            rows = crsr.fetchone()
            crsr.close()
            conn.close()
            return rows
        except Exception as ex:
            conn.close()
            return None

    def execute_query(self, query):
        conn = self.connect()
        try:
            # query = query.replace("/", "-")
            crsr = conn.cursor()
            crsr.execute(query)
            conn.commit()
            crsr.close()
            conn.close()
            return True
        except Exception as ex:
            conn.close()
            print(ex) 
            print(query)
            return False

    def execute_query_with_params(self, query, params):
        conn = self.connect()
        try:
            crsr = conn.cursor()
            crsr.execute(query, params)
            conn.commit()
            crsr.close()
            conn.close()
            return True
        except Exception as ex:
            conn.close()
            return False
        
    def check_record_exist(self, query):
        conn = self.connect()
        try:
            crsr = conn.cursor()
            crsr.execute(query)
            rows = crsr.fetchone()
            crsr.close()
            conn.close()
            if rows is not None:
                return True
            return False
        except Exception as ex:
            conn.close()
            return False
        
    def check_record_exist_with_params(self, query, params):
        conn = self.connect()
        try:
            crsr = conn.cursor()
            crsr.execute(query, params)
            rows = crsr.fetchone()
            crsr.close()
            conn.close()
            if rows is not None:
                return True
            return False
        except Exception as ex:
            conn.close()
            return False
```
